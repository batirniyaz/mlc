# Python Basics

## In python there are variance of data types.
### Because of dynamic python we can store any kind of data to iteratable data types like list, dict and etc.

### List (Array)
defines with square brackets

```python
my_list = [1, 2, 'three', True]
type(my_list)
```

### Dictionary (key-value)
defines with figure brackets and achieve key-value throught colon

```python
my_dict = {'a': 'A value 1', 'b': 'A value 2', 'c': my_list}
type(my_dict)
```

### Set
defines with figure brackets and there values are unique

```python
my_set = {'a', 'b', 'c', 2, 3}
type(my_set)
```

### Tuple
defines with parentheses and inside values cannot be changed

```python
my_tuple = (3, 4, 6, 'a', 'g')
type(my_tuple)
```

## Simple calculate circle area program

### The formula of finding the area of circle is (A = Pi * r * r)

```python
radius = int(input("Enter the radius of circle: "))

area = 3.14 * radius ** 2
print(area)
```
