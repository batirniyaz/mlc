# Computer Vision - Manipulation with Image

````python
# Import necessary libs
import cv2
import matplotlib.pyplot as plt
import numpy as np
````

````python
img_path = 'cat_experiment.jpg'

# Load the image with OpenCV
img = cv2.imread(img_path)
````

Firstly, I will convert it from BGR to RGB since OpenCV loads images in BGR order to see the original image

````python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
````

````python
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
````

### Converting to Grayscale

````python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
````

### Applying Blurring

````python
mean_filtered = cv2.blur(img_gray, (5, 5))

gaussian_filtered = cv2.GaussianBlur(img_gray, (7, 7), 0)
````

````python
# Display the original image, mean filtered image, and Gaussian blurred image
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filter (5x5)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Blur (7x7)')
plt.axis('off')

plt.tight_layout()
plt.show()
````

I increased the kernel size of mean filter to 5x5 and gaussian blur to 7x7 because of no changes in 3x3 or just minimum changes

### Applying Sharpening

````python
# Define the sharpening kernel
sharpening_kernel = np.array([[ 0, -1,  0],
                              [-1,  5, -1],
                              [ 0, -1,  0]])

sharpened = cv2.filter2D(img_gray, -1, sharpening_kernel)
````

````python
# Display the original grayscale image and the sharpened image side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
````

### Applying Edge Detection

Sobel Edge Detection

````python
# Calculate the Sobel gradient in the x and y direction
sobel_x = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)

# Combine the gradients by calculating the gradient magnitude and normilize it
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_magnitude = np.uint8(255 * sobel_magnitude / np.max(sobel_magnitude))
````

Canny Edge Detection

````python
# Threshold values can be adjusted depending on the image
canny_edges = cv2.Canny(img_gray, 100, 200)
````

````python
# Display the images side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
````

The Sobel filter is applied separately in the x and y directions. Canny Edge Detection is applied with threshold values 100 and 200.
